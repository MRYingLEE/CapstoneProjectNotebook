```
import pandas as pd

df_toronto=pd.read_csv("toronto.csv",index_col="PostalCode")
df_toronto.head()
```

```
df_toronto.info()
```

```
df_toronto=df_toronto.loc[:,df_toronto.columns[1:]]
df_toronto.head()
```

```
df_geo=pd.read_csv("https://cocl.us/Geospatial_data",index_col="Postal Code")
```

```
df_geo.index.rename('PostalCode',inplace=True)
df_geo.head()
```

```
df_geo.columns
```

```
df_toronto=df_toronto.join(df_geo, how="left")
```

```
df_toronto.head()
```

```
df_toronto.to_csv("toronto_ext.csv")
```
