```

import pandas as pd
import requests
```

# Load Data

```
wiki = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
wikipedia_page = requests.get(wiki)

df_web = pd.read_html(wikipedia_page.content, header=0)[0]
df_web.head()
```

# Clean Data

```
df_clean=df_web[df_web["Borough"]!="Not assigned"]
df_clean.head(10)
```

```

df_clean.loc[df_clean.Neighbourhood == 'Not assigned']
```

```
df_clean.Neighbourhood.replace('Not assigned',df_clean.Borough,inplace=True)
```

```
df_clean.head(10)
```

```
df_toronto = df_clean.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(lambda x: ', '.join(x))
df_toronto = df_toronto.reset_index()
df_toronto.rename(columns = {'Postcode':'PostalCode'}, inplace = True)
df_toronto.rename(columns = {'Neighbourhood':'Neighborhood'}, inplace = True)
df_toronto.head(10)
```

```
df_toronto.set_index("PostalCode")
```

```
df_toronto.to_csv("toronto.csv",sep="|" ,encoding="utf-16")

from google.colab import files


files.download('toronto.csv')
```
